```python
img1 = cv2.imread('/content/20170617_stp003.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img1.shape
```

```python
all_pixels= img1.reshape(720*1280, 3)
```

```python
all_pixels.shape
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=10, n_jobs=-1)
```

```python
km.fit(all_pixels)
```

```python
centres = km.cluster_centers_
```

```python
centres
```

```python
centres = np.array(centres, dtype='uint')
```

```python
centres
```

```python
km.labels_.shape
```

```python
np.unique(km.labels_, return_counts = True)
```

```python
new_img = np.zeros((all_pixels.shape[0],3), dtype= 'uint')
```

```python
plt.imshow(new_img.reshape(720,1280, 3))
```

```python
for i in range(new_img.shape[0]):
  new_img[i] = centres[km.labels_[i]]
  
```

```python
# n_clusters = 5
plt.imshow(new_img.reshape(720,1280,3))
plt.show()
```
